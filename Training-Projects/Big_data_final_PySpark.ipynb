# 8. Загрузка данных

```
from pyspark.sql.types import StructType, StructField, IntegerType

schema = StructType([
    StructField("user_id", IntegerType(), True),
    StructField("item_id", IntegerType(), True),
    StructField("rating", IntegerType(), True),
    StructField("timestamp", IntegerType(), True)])

df_ratings = spark.read.csv('u.data.csv', header=False, schema=schema, sep = '\t')
```

```
df_ratings.show(10)
```

```
df_movies = spark.read.csv('u.item.csv', header=False, sep='|')
```

```
df_movies = df_movies.withColumnRenamed("_c0","movie_id").withColumnRenamed("_c1","movie_title").withColumnRenamed("_c2","release_date").withColumnRenamed("_c3","video_release_date").withColumnRenamed("_c4","imdb_url").withColumnRenamed("_c5","unknown").withColumnRenamed("_c6","action").withColumnRenamed("_c7","adventure").withColumnRenamed("_c8","animation").withColumnRenamed("_c9","childrens").withColumnRenamed("_c10","comedy").withColumnRenamed("_c11","crime").withColumnRenamed("_c12","documentary").withColumnRenamed("_c13","drama").withColumnRenamed("_c14","fantasy").withColumnRenamed("_c15","film_noir").withColumnRenamed("_c16","horror").withColumnRenamed("_c17","musical").withColumnRenamed("_c18","mystery").withColumnRenamed("_c19","romance").withColumnRenamed("_c20","sci_fi").withColumnRenamed("_c21","thriller").withColumnRenamed("_c22","war").withColumnRenamed("_c23","western")
```

```
df_movies.show(10)
```

# 9. Выводим среднюю оценку для каждого фильма

```
from pyspark.sql.functions import *
```

```
# объединяем датафреймы
new_df = df_ratings.join(df_movies, df_ratings.item_id == df_movies.movie_id).select(df_ratings["*"],df_movies["*"])
```

```
new_df.show(5)
```

```
from pyspark.sql import functions as F
```

```
# средний рейтинг для каждого фильма
avg_film_rating = new_df.groupBy("item_id").agg(F.mean('rating'))
avg_film_rating.show()
```

# 10. Выводим среднюю оценку для каждого жанра

```
grouped_table = new_df.groupBy('movie_id').avg()
grouped_table = grouped_table.withColumnRenamed("avg(rating)","avg_rating")
grouped_table.show(5)
```

```
final_table = grouped_table.join(df_movies, grouped_table.movie_id == df_movies.movie_id).select(grouped_table['avg_rating'],df_movies["*"])
final_table.show(5)
```

```
col_list = new_df.columns
genres = col_list[9:]
```

```
for genre_name in genres:
  print(genre_name)
  genre_avg_rating = final_table.where(final_table[genre_name]==1)
  genre_avg_rating = genre_avg_rating.groupBy([genre_name]).avg()
  genre_avg_rating.show()
```

```
new_df.createOrReplaceTempView('genre_avg')
```

```
# второй способ первая часть (не актуально)

first_table = spark.sql('\
SELECT avg(rating) as avg_rating, unknown, action, adventure, animation, childrens, comedy, crime, documentary, drama, fantasy \
FROM genre_avg \
WHERE unknown=1 or action=1 or adventure=1 or animation=1 or childrens=1 or comedy=1 or crime=1 or documentary=1 or drama=1 or fantasy=1 \
GROUP BY unknown, action, adventure, animation, childrens, comedy, crime, documentary, drama, fantasy \
ORDER BY avg_rating desc \
')
```

```
# второй способ вторая часть (не актуально)

second_table = spark.sql('\
SELECT distinct avg(rating) as avg_rating, film_noir, horror, musical, mystery, romance, sci_fi, thriller, war, western \
FROM genre_avg \
WHERE film_noir=1 or horror=1 or musical=1 or mystery=1 or romance=1 or sci_fi=1 or thriller=1 or war=1 or western=1 \
GROUP BY film_noir, horror, musical, mystery, romance, sci_fi, thriller, war, western \
ORDER BY avg_rating desc\
')
```

# 11. Получаем 2 датафрейма с 5-ю самыми популярными и самыми непопулярными фильмами (по средней оценке)

```
avg_film_rating = avg_film_rating.withColumnRenamed("avg(rating)","avg_rating")
```

```
avg_film_rating.show(10)
```

```
avg_film_rating.createOrReplaceTempView('film_rating')
```

```
# топ-5

top_rating = spark.sql('\
SELECT * \
FROM film_rating \
ORDER BY avg_rating DESC \
LIMIT 5\
')
top_rating.show()
```

```
# топ-5 снизу (антирейтинг)

top_antirating = spark.sql('\
SELECT * \
FROM film_rating \
ORDER BY avg_rating ASC \
LIMIT 5\
')
top_antirating.show()
```
